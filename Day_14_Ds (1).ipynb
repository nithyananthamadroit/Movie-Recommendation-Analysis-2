# Classification

1. Loan Approval Prediction

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data = {
    'income': [25, 30, 35, 40, 45, 50, 55, 60],
    'credit_score': [550, 580, 600, 650, 680, 700, 720, 750],
    'loan_approved': [0, 0, 0, 1, 1, 1, 1, 1]
}

loan_data = pd.DataFrame(data)
loan_data.head()
```

```python
loan_input_data = loan_data[['income','credit_score']]
loan_ouput_data = loan_data['loan_approved']
```

```python
x_train,x_test,y_train,y_test=train_test_split(loan_input_data,loan_ouput_data,test_size=0.30,random_state=42)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
predicted_output = model.predict(x_test)
predicted_output
```

```python
final_score = accuracy_score(y_test,predicted_output)
print("Accuracy Score is :",final_score)
```

```python
detailed_score = confusion_matrix(y_test,predicted_output)
print(detailed_score)
```

```python
salary = float(input("Enter you salary amount(In Thousands): "))
credit_score = int(input("Enter your credit Score: "))

new_applicant = [[salary,credit_score]]

if(new_applicant[0][0] <=0 or new_applicant[0][1] > 900):
  print("Invalid Values")
elif(new_applicant[0][0] < 40000):
  print("Application Rejected Due to Low Salary")
elif(new_applicant[0][1] <= 600):
  print("Loan not approved due to low credit score")
else:
  prediction = model.predict(new_applicant)
  if(prediction[0] == 1):
    print("The Final Result of your Loan Application is Approved")
  else:
    print("The Final Result of your Loan Application is Not Approved")
```

Rules
  1. Salary should not be negative
  2. credit score should not be more than 900
  3. Salary should be minimum of 40k
  4. credit score should not be less than 600
